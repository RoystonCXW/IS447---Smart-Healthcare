# Add additional metric

### Overall Time Taken to Dispensed

```python
# handle Unknown string format: 9.:09 in Time TTO reached
df2['Time TTO Reached'] = df2['Time TTO Reached'].replace('9.:09', 'NA')
```

```python
tto_received = pd.to_datetime(df2['Time TTO Received'].astype(str)) 
tto_reached = pd.to_datetime(df2[df2['Time TTO Reached'] != 'NA']['Time TTO Reached'].astype(str))

df2['Overall Time Taken'] = tto_reached.sub(tto_received).dt.total_seconds().div(60)
```

### Add whether meet KPI column

```python
def f(row):
    if row['Overall Time Taken'] > 45:
        val = 0
    elif row['Overall Time Taken'] < 0:
        val = 'NA'
    else:
        val = 1
    return val
```

```python
df2['Meet KPI'] = df2.apply(f, axis=1)
```
